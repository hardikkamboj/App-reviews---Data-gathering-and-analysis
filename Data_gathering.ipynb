```python
def get_reviews(link,output_file,return_file = False):
  from google_play_scraper import Sort, reviews_all
  import pandas as pd
  from tqdm import tqdm

  result = reviews_all(
    link,
    sleep_milliseconds=0, # defaults to 0
    lang='en', # defaults to 'en'
    country='us', # defaults to 'us'
    sort=Sort.MOST_RELEVANT, # defaults to Sort.MOST_RELEVANT
    filter_score_with=None # defaults to None(means all score)
      )

  df = pd.DataFrame()
  for i in tqdm(range(len(result))):
    row = pd.DataFrame(result[i],index = [i])
    df = pd.concat([df,row],axis=0)

  df.to_csv(output_file)

  if return_file:
     return df
```

```python
app_info = {'axis_merchant':'com.axismerchant',
            'hdfc_smarthub':'com.ionicframework.payzaap483385',
            'icici_eazypay':'com.icici.eazypaycollect',
            'alltap':'com.pinelabs.epos',
            'payswift':'com.payswiff.set.app'}
```

```python
for app in app_info:
  name = app + '.csv'
  get_reviews(app_info[app],name)
```
